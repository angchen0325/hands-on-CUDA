$\texttt{Q1}$. There are three `for` statements in this chapter's Mandelbrot example; however, we can only parallelize over the first two. 
Why can't we parallelize over all of the for loops here?

$\texttt{A1}$. The first two `for` loops iterate over every pixel, whose outputs are invariant to each other;
we can thus parallelize over these two `for` loops.
The third `for` loop calculates the final value of a particular pixel, which is intrinsically recursive.

$\texttt{Q2}$. What is something that Amdahl's Law doesn't account for when we apply it to offloading a serial CPU algorithm to a GPU?

$\texttt{A2}$. Amdahl's Law doesn't account for the time it takes to transfer memory between the GPU and the host.

$\texttt{Q3}$. Suppose that you gain exclusive access to three top-secret GPUs that are the same in all respects, except for core counts -- the first has 131,072 cores, the second has 262,144 cores, and the third has 524,288 cores. 
If you parallelize and offload the Mandelbrot example onto these GPUS (which generates a 512x512 pixel image), will there be a difference in computation time between the first and second GPU?
How about between the second and third GPU?

$\texttt{A3}$.  512x512 amounts to 262,144 pixels.
This means that the first GPU can only calculate the outputs of half of all the pixels, while the second GPU can calculate all of the pixels at once; this means that the second GPU will be about twice as fast as the first GPU.
The third GPU has more than sufficient cores ot calculate all pixels at once.
We have known that the extra cores will be of no use to us. 
And thus the second and third GPUS will be equally fast for this problem.

$\texttt{Q4}$. Can you think of any problems with designating certain algorithms or blocks of code as *parallelizable* in the context of Amdahl's Law?

$\texttt{A4}$. If so, the computation time for this piece of code will be close to *0* if the number of processors, $N$, is very large.
As we can see from the last problem, this is not the case.

$\texttt{Q5}$. Why should we use profilers instead of just using Python's *time* function?

$\texttt{A5}$. First, using *time* consistently can be cumbersome, and it might not zero in on the bottlenecks of your program.
Second, a profiler can tell you the exact computation time of all of your code from the perspective of Python, so you can tell whether some library function or background activity of your operating system is at fault rather than your code.
