$\texttt{Q1}$. There are three `for` statements in this chapter's Mandelbrot example; however, we can only parallelize over the first two. 
Why can't we parallelize over all of the for loops here?

$\texttt{A1}$. The first two `for` loops iterate over every pixel, whose outputs are invariant to each other;
we can thus parallelize over these two `for` loops.
The third `for` loop calculates the final value of a particular pixel, which is intrinsically recursive.
