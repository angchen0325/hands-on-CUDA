$\texttt{Q1}$. Can we run CUDA on our main processor's built-in Intel HD GPU? 
What about on discrete AMD Radeon GPU?

$\texttt{A1}$. No, CUDA only supports NVIDIA GPUs, not Intel HD or AMD Radeon.
